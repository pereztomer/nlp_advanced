```python
import pandas as pd
```

```python
df = pd.read_csv('embrassing_400_by_annotators.csv')
df
```

```python
df_filtered = df.dropna(subset=['Tomer'])
df_filtered = df_filtered[['example_id', 'preprocessed_text', 'Tomer']]
df_filtered
```

```python
# Calculate label distribution
label_distribution = df_filtered['Tomer'].value_counts()

# Print the label distribution
print("Label Distribution:")
print(label_distribution)

# Calculate general statistics on labels
label_stats = df_filtered['Tomer'].describe()

# Print the label statistics
print("\nLabel Statistics:")
print(label_stats)
```
